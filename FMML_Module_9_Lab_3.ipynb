Central to CNNs, a convolution operation is a linear element-wise multiplication operation between a small filter/kernel and same-sized patch from the image. We move this filter over the image like a sliding window from top left to bottom right. For each point on the image, a value is calculated based on the filter using a convolution operation. These filters can do simplest task like checking if there is a vertical line in the image or complicated task like detecting a human eye in the image.

Let's look at the convolution formula:

Convolution between image
$f(x, y)$ and kernel $k(x, y)$ is
$$f(x,y) * k(x,y) = \sum \limits _{i=0} ^{W-1} \sum \limits _{j=0} ^{H-1} f(i, j) k(x − i, y − j)$$

where $W$ and $H$ are the the width and height of the image.

```python
# 2D 3x3 binary image with vertical edge
image1 = np.array([[1,1,0], [1,1,0], [1,1,0]])

# 2D 3x3 binary image with horizontal edge
image2 = np.array([[0,0,0], [0,0,0], [1,1,1]])

# On plotting the images
plt.imshow(image1, cmap='gray', extent=[0, 3, 3, 0])
plt.show()
plt.imshow(image2, cmap='gray', extent=[0, 3, 3, 0])
plt.show()
```

```python
# Vertical Line filter
filter = np.array([[1,0,-1],
                   [1,0,-1],
                   [1,0,-1]])

# Applying filter to first image
output = np.sum(np.multiply(image1, filter))
print('Output from first image: ', output)

# Applying filter to second image
output = np.sum(np.multiply(image2, filter))
print('Output from second image: ', output)
```

```python
# Horizontal edge filter
filter = np.array([[-1,-1,-1],
                   [ 0, 0, 0],
                   [ 1, 1, 1]])

output = np.sum(np.multiply(image1, filter))
print('Output from first image: ', output)

output = np.sum(np.multiply(image2, filter))
print('Output from second image: ', output)
```

Non-zero output suggests that there is a vertical edge present in the first image and not present in the second image. Similarly, horizontal edge is detected in second.

Let's define a function to use convolution layer from Pytorch and use our own kernel to detect edges in image

```python
def apply_conv(image, kernel, padding=0, stride=1):

  #--------IMAGE PREPROCESSING-------
  image = torch.from_numpy(image)
  # Pytorch requires input to convolution in (N,C,H,W), where N = batch size and C=#channels in input
  input = image.view((1,1,image.shape[0], image.shape[1]))

  # --------------KERNEL-------------
  kernel = torch.from_numpy(kernel.astype(np.float32))

  # Pytorch requires kernel of shape (N,C,H,W), where N = batch size and C=#channels in input
  kernel = kernel.view((1,1,kernel.shape[0], kernel.shape[1]))

  # ---------CONVOLUTION LAYER from Pytorch--------
  conv = Conv2d(in_channels=1, out_channels=1, kernel_size=kernel.shape, padding=padding, stride=stride)

  # Set the kernel weights in the convolution layer
  conv.weight = torch.nn.Parameter(kernel)

  # ---------APPLY CONVOLUTION--------
  output = conv(input.float())
  output_img = output.data.numpy()  # Tensor to back in numpy
  output_img = output_img.reshape((-1, output_img.shape[-1])) # Reshape to 2D image

  return output_img
```

```python
# Our original lotus image
image = cv2.imread('/content/grid1.jpg', 0)

filter = np.array([[-1,-1,-1],
                   [ 0, 0, 0],
                   [ 1, 1, 1]])

out1 = apply_conv(image, filter, padding=0, stride=1)

filter = np.array([[1,0,-1],
                   [1,0,-1],
                   [1,0,-1]])

out2 = apply_conv(image, filter, padding=0, stride=1)
```

## Softmax layer/activation
Recall that logistic regression produces a decimal between 0 and 1.0. For example, a logistic regression output of 0.8 from an email classifier suggests an 80% chance of an email being spam and a 20% chance of it being not spam. Clearly, the sum of the probabilities of an email being either spam or not spam is 1.0.

Softmax extends this idea into a multi-class world. That is, Softmax assigns decimal probabilities to each class in a multi-class problem. Those decimal probabilities must add up to 1.0. This additional constraint helps training converge more quickly than it otherwise would.
Softmax is implemented through a neural network layer just before the output layer. The Softmax layer must have the same number of nodes as the output layer.

## Learning to train a CNN network

```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  Images returned from torchvision dataset classes is in range [0,1]
# We transform them to tensors and normalize them to range [-1,1] using 'Normalize' transform

transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

batch_size = 128

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)

# Classes in CIFAR10
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
print('Training data shape : ', trainset.data.shape, len(trainset.targets))
print('Testing data shape : ', testset.data.shape, len(testset.targets))

# Find the unique numbers from the train labels
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

```python
def train(num_epochs, model, train_loader, loss_func, optimizer):

  # Training mode
  model.train()

  train_losses = []
  train_acc = []

  # Train the model
  for epoch in range(num_epochs):
    running_loss = 0
    running_acc = 0
    for i, (images, labels) in enumerate(train_loader):

      # clear gradients for this training step
      optimizer.zero_grad()

      # Forward pass
      output = model(images)

      # Calculate loss
      loss = loss_func(output, labels)

      # Backpropagation, compute gradients
      loss.backward()

      # Apply gradients
      optimizer.step()

      # Running loss
      running_loss += loss.item()

      # indices of max probabilities
      _, preds = torch.max(output, dim=1)

      # Calculate number of correct predictions
      correct = (preds.float() == labels).sum()
      running_acc += correct

      # Average loss and acc values
      epoch_loss = running_loss / len(train_loader.dataset)
      epoch_acc = running_acc / len(train_loader.dataset)

    train_losses.append(epoch_loss)
    train_acc.append(epoch_acc)
    print ('Epoch {}/{}, Loss: {:.4f}, Accuracy: {:.4f}'.format(epoch + 1, num_epochs, epoch_loss, epoch_acc*100))

  return train_losses, train_acc
```

```python
def test_model(model, testloader):
  correct = 0
  total = 0
  # Deactivate autograd engine (don't compute grads since we're not training)
  with torch.no_grad():
    for data in testloader:
        images, labels = data
        # Calculate outputs by running images through the network
        outputs = model(images)
        # The class with the highest value is what we choose as prediction
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

  print('Accuracy of the network: %d %%' % (
      100 * correct / total))
```

```python
# CNN with 2 CONV layers and 3 FC layers
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, 3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, 5)
        self.fc1 = nn.Linear(32 * 5 * 5, 512)
        self.fc2 = nn.Linear(512, 128)
        # output layer 10 classes
        self.fc3 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        # flatten all dimensions except batch
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
model = Net()
print(model)
```

```python
# Cross Entropy loss for multi-class classification
criterion = nn.CrossEntropyLoss()
```

```python
# SGD optimizer with momentum
optimizer = optim.SGD(model.parameters(), lr=0.05, momentum=0.9)
```

```python
num_epochs = 5  # iterations
train_losses, train_acc = train(num_epochs, model, trainloader, criterion, optimizer)
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(1,2, 1)
ax.plot(np.arange(1,len(train_losses)+1),train_losses)
plt.xlabel('Training loss')
plt.ylabel('Epochs')
ax.set_title('Loss vs Epochs')
ax = fig.add_subplot(1,2, 2)
ax.plot(np.arange(1,len(train_acc)+1),train_acc)
plt.xlabel('Training accuracy')
plt.ylabel('Epochs')
ax.set_title('Accuracy vs Epochs')
plt.show()
```

```python
# Accuracy on test data after training
test_model(model, testloader)
```

#QUESTION-1


Convolutional Neural Networks (CNNs) are generally preferred over traditional Artificial Neural Networks (ANNs) for image classification tasks due to several reasons:

1. **Hierarchical Feature Learning**: CNNs are designed to automatically learn hierarchical representations of features from raw pixel values. They use convolutional layers to capture local patterns, followed by pooling layers to aggregate information hierarchically. This hierarchical feature learning is crucial for capturing spatial dependencies in images, which ANNs may struggle with.

2. **Parameter Sharing**: CNNs exploit the spatial locality of patterns in images by using parameter sharing in convolutional layers. This drastically reduces the number of parameters compared to ANNs, making CNNs more efficient to train and less prone to overfitting, especially on large datasets.

3. **Translation Invariance**: CNNs inherently possess translation invariance, meaning they can recognize patterns regardless of their position in the image. This property is crucial for tasks like object recognition, where the position of an object within an image may vary.

4. **Feature Reuse**: CNN architectures often include multiple layers of convolution and pooling, which allow for the reuse of learned features across different parts of the image. This promotes generalization and robustness, as features learned in early layers (e.g., edges, textures) can be useful for recognizing different objects in later layers.

5. **Local Connectivity**: CNNs exploit the local connectivity present in images. Instead of connecting every neuron in one layer to every neuron in the next layer (as in fully connected ANNs), CNNs connect each neuron to only a local region of the input volume. This local connectivity preserves spatial information and reduces the computational cost.

6. **Regularization Techniques**: CNN architectures often incorporate regularization techniques like dropout and batch normalization, which help prevent overfitting and improve generalization performance.

7. **Pretrained Models and Transfer Learning**: CNNs trained on large datasets like ImageNet are readily available, and their pretrained weights can be fine-tuned on specific image classification tasks with smaller datasets. This transfer learning approach allows for efficient training even with limited labeled data.

Overall, CNNs are specifically tailored for processing visual data like images, making them more effective and efficient for image classification tasks compared to traditional ANNs.

Reducing the number of convolutional layers to only one in a Convolutional Neural Network (CNN) simplifies the network architecture but may also lead to a decrease in performance, especially for complex datasets. Here's what happens when you reduce the number of convolutional layers to one:

1. **Decreased Model Complexity**: With only one convolutional layer, the model becomes less complex. This can reduce the number of parameters and computational cost, making the model faster to train and evaluate.

2. **Limited Feature Extraction**: Convolutional layers are responsible for extracting features from the input images. With only one layer, the network has limited capacity to learn complex hierarchical features. As a result, it may struggle to capture intricate patterns in the data.

3. **Loss of Hierarchical Representation**: Deep CNN architectures typically learn hierarchical representations of features, where lower layers capture low-level features like edges and textures, and higher layers capture more abstract features like object parts and shapes. With only one convolutional layer, the network may not be able to learn these hierarchical representations effectively.

4. **Reduced Model Performance**: Removing convolutional layers can lead to a decrease in model performance, especially on tasks requiring high-level feature understanding, such as image classification of complex datasets like ImageNet. The model may struggle to generalize well to unseen data and may suffer from overfitting.

5. **Sensitivity to Image Variability**: CNNs with fewer convolutional layers may be more sensitive to variations in input images, such as changes in scale, rotation, or illumination. They may not be able to learn robust features that are invariant to these transformations.

However, reducing the number of convolutional layers to one might still be appropriate for simpler datasets or tasks where the features are relatively easy to learn. It can also be beneficial in scenarios where computational resources are limited, and a simpler model is preferred. Overall, the impact of reducing the number of convolutional layers depends on the specific dataset, task, and performance requirements.

The Softmax activation function is typically used in the last layer of a Convolutional Neural Network (CNN) when the task is multi-class classification, where each input can belong to only one class. It converts the raw scores (logits) produced by the network into probabilities, ensuring that the output values are non-negative and sum up to 1, thus representing class probabilities.

However, there are scenarios where the Softmax activation function may not be used in the last layer of a CNN:

1. **Binary Classification**: If the task is binary classification (i.e., classifying inputs into two classes), it's common to use a single neuron in the output layer with a sigmoid activation function instead of multiple neurons with Softmax. The sigmoid function squashes the output to the range [0, 1], representing the probability of the input belonging to one of the two classes.

2. **Regression**: In regression tasks, where the goal is to predict a continuous value rather than class probabilities, the last layer may have a different activation function such as linear or ReLU, depending on the nature of the output.

3. **Custom Output Layer**: In some cases, especially when dealing with specialized architectures or custom loss functions, a different activation function or no activation function at all may be used in the last layer of the CNN.

It's important to choose the appropriate activation function based on the task and the desired properties of the output. While Softmax is commonly used for multi-class classification tasks, its usage may vary depending on the specific requirements of the problem at hand.
